```python
import pandas as pd
df1=pd.read_csv("/content/Depression_BERT.csv")
df2=pd.read_csv("/content/Depression_RoBERTa.csv")
df3=pd.read_csv("/content/Depression_BERT.csv")
```

```python
l1=df1['Labels']
l2=df2['Labels']
l3=df3['Labels']
```

```python
l4=[]
```

```python
for i in range (len(l1)):
  c1=0
  c2=0
  c3=0
  if l1[i]=='not depression':
    c1+=1 
  elif l1[i]=='moderate':
    c2+=1
  else:
    c3+=1
  if l2[i]=='not depression':
    c1+=1 
  elif l2[i]=='moderate':
    c2+=1
  else:
    c3+=1
  if l3[i]=='not depression':
    c1+=1 
  elif l3[i]=='moderate':
    c2+=1
  else:
    c3+=1
  if c1>=2:
    l4.append('not depression')
  elif c2>=2:
    l4.append("moderate")
  else c3>=2:
    l4.append("severe")
```
